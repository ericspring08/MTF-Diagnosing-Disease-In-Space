```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('./Experiments/10,000_Trials_#1/stats.csv')
```

```python
for i in df.columns:
    if min(df[i]) != 1.0 and ("accuracy" in i or "precision" in i) and 'balanced' not in i and 'rwma' in i:
        plt.hist(df[i])
        plt.title(i)
        plt.ylabel('Number of Trials')
        plt.show()
```

```python
for i in df.columns:
    if min(df[i]) != 1.0 and 'time to fit' in i and 'lcx' in i:
        plt.hist(df[i])
        plt.title(i)
        plt.ylabel('Number of Trials')
        plt.show()
```
